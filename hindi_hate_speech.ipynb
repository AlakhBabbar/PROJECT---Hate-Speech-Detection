```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Hinglish_Profanity_List.csv", encoding="latin")
df.head()
```

```python
def simplify(text):
    vowels = "aeiouAEIOU"
    for vowel in vowels:
      text = text.replace(vowel, "")
    return text
```

```python
df['Hindi']=df['Hindi'].apply(simplify)
```

```python
df.head(50)
```

```python
def hate(text):
  text = text.lower()
  text = text.split(" ")
  text = [simplify(word) for word in text]
  print(text)
  for i in text:
    if i in df.values:
      print(i)
      return True
      break
  return False
```
